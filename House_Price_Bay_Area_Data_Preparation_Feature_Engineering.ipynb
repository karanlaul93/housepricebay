```python
#Generating p values using ANOVA test
import scipy.stats
cat_features=cat_data1.columns
df_train1[cat_features]=df_train1[cat_features].fillna('Missing')
```

```python
anova={'feature':[],'f':[],'p':[]}
for cat in cat_features:
    group_prices=[]
    for group in df_train1[cat].unique():
        group_prices.append(df_train1[df_train1[cat]==group]['SalePrice'].values)
    f,p=scipy.stats.f_oneway(*group_prices)
    anova['feature'].append(cat)
    anova['f'].append(f)
    anova['p'].append(p)
anova=pd.DataFrame(anova)
anova=anova[['feature','f','p']]
anova.sort_values('p',inplace=True)
    
```

```python
#If our p value is <0.05 we reject our null hypothesis
#Null hypothesis is, all the above categorical features, do not have any huge difference (same) 
#or high importance in contributing in the sale price

#If p < 0.05 we reject this null hypothesis that means that features has importance in contributing in the sale price
```

```python
# # #Standardizing the numerical features (mean 0 and SD 1)
from sklearn.preprocessing import StandardScaler

num_features1.remove('Id')
num_features1.remove('SalePrice')
stdsc=StandardScaler()
X_train.loc[:,num_features1]=stdsc.fit_transform(X_train[num_features1])
X_test.loc[:,num_features1]=stdsc.fit_transform(X_test[num_features1])
```

```python
#Defining cost function(root sq mean error)
from sklearn.model_selection import cross_val_score

def rsme(model,X,y):
    cv_scores=-cross_val_score(model,X,y,scoring="neg_mean_squared_error",cv=10)
    return np.sqrt(cv_scores)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

param_grid={'alpha':[0.0001,0.001,0.01,0.1,1.,5.,10.,25.],'max_iter':[5000]}
lasso=GridSearchCV(Lasso(),cv=5,param_grid=param_grid,scoring="neg_mean_squared_error")
lasso.fit(X_train,y_train)
alpha=lasso.best_params_['alpha']

#Hone in
param_grid={'alpha':[x/100. * alpha for x in range (50,150,5)],'max_iter':[5000]}
lasso=GridSearchCV(Lasso(),cv=5,param_grid=param_grid,scoring="neg_mean_squared_error")
lasso.fit(X_train,y_train)
alpha=lasso.best_params_['alpha']
lasso=lasso.best_estimator_

print("Lasso -- Train RSME: {:.5f} | Test RSME: {:.5f} | alpha: {:.5f}".format(rsme(lasso,X_train,y_train).mean(),rsme(lasso,X_test,y_test).mean(),alpha))
```

```python
#Removing features which are not affecting using Lasso methodology
import seaborn as sns
coefs=pd.DataFrame({'coefs':lasso.coef_,'Positive':lasso.coef_>0},index=X_train.columns)
coefs['coefs_abs']=np.abs(coefs.coefs)
print("LASSO dropped {} of {} features".format(sum(coefs.coefs==0),coefs.shape[0]))

top_coefs=coefs.sort_values('coefs_abs',ascending=False).head(20)
plt.figure(figsize=(8,10))
sns.barplot(top_coefs.coefs_abs,top_coefs.index,orient='h',hue=top_coefs.Positive)
plt.title("LASSO Regression: Top Features")
plt.xlabel("Absolute Coeficient")
plt.show()
```
